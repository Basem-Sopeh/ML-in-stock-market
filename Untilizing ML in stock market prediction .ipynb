```python
# get_stats func gives us the statistics of any strategy 
# including wins, losses, means, standard diviations, and the sharp ratio
# sharp ratio is a quantity that evaluates the return ofn an investment given a certain risk or volatility
```

```python
def get_stats(s, n=252):
    s = s.dropna()
    wins = len(s[s>0])
    losses = len(s[s<0])
    evens = len(s[s==0])
    mean_w = round(s[s>0].mean(), 3)
    mean_l = round(s[s<0].mean(), 3)
    win_r = round(wins/losses, 3)
    mean_trd = round(s.mean(), 3)
    sd = round(np.std(s), 3)
    max_l = round(s.min(), 3)
    max_w = round(s.max(), 3)
    sharpe_r = round((s.mean()/np.std(s))*np.sqrt(n), 4)
    cnt = len(s)
    print('Trades:', cnt,\
        '\nWins:', wins,\
        '\nLosses:', losses,\
        '\nBreakeven:', evens,\
        '\nWin/Loss Ratio', win_r,\
        '\nMean Win:', mean_w,\
        '\nMean Loss:', mean_l,\
        '\nMean', mean_trd,\
        '\nStd Dev:', sd,\
        '\nMax Loss:', max_l,\
        '\nMax Win:', max_w,\
        '\nSharpe Ratio:', sharpe_r)
```

```python
# Building a model and evaluation the performance
# arranging the data to predict the close price of next da depending on the prices of the prior 20 days
```

```python
for i in range(1, 21, 1):
    sp.loc[:,'Close Minus ' + str(i)] = sp['Close'].shift(i)
    sp20 = sp[[x for x in sp.columns if 'Close Minus' in x or x =='Close']].iloc[20:,]
sp20
```

```python
# reversing the dataframe
# the close price of the next day will be at the last column
sp20 = sp20.iloc[:,::-1]
sp20
```

```python
# applying support vector machine algorithm
# the last 1000 rows of the model are used as test data
# the rest are used as training data
```

```python
from sklearn.svm import SVR
clf = SVR(kernel='linear')
X_train = sp20[:-1000]
y_train = sp20['Close'].shift(-1)[:-1000]
X_test = sp20[-1000:]
y_test = sp20['Close'].shift(-1)[-1000:]
```

```python
sp20['Close'][-1000:]
```

```python
# fitting thr model & applying it SVR to predictthe test data
model = clf.fit(X_train, y_train)
preds = model.predict(X_test)
```

```python
# test dataframe
tf = pd.DataFrame(list(zip(y_test, preds)), columns=['Next Day Close','Predicted Next Close'], index=y_test.index)
tf
```

```python
# the scenario: buying the next day open if the predicted next close is higher than the open price
cdc = sp[['Close']].iloc[-1000:]
ndo = sp[['Open']].iloc[-1000:].shift(-1)
tf1 = pd.merge(tf, cdc, left_index=True, right_index=True)
tf2 = pd.merge(tf1, ndo, left_index=True, right_index=True)
tf2.columns = ['Next Day Close', 'Predicted Next Close', 'Current Day Close', 'Next Day Open']
tf2
```

```python
# checking the mentioned condition required to buy the stock
def get_signal(r):
    if r['Predicted Next Close'] > r['Next Day Open']:
        return 1
    else:
        return 0
```

```python
# computing the real return _ in point_
def get_ret(r):
    if r['Signal'] == 1:
        return ((r['Next Day Close'] - r['Next Day Open'])/r['Next Day Open']) * 100
    else:
        return 0
```

```python
tf2 = tf2.assign(Signal = tf2.apply(get_signal, axis=1))
tf2 = tf2.assign(PnL = tf2.apply(get_ret, axis=1))
tf2
```

```python
# the return if we follow this strategy
(tf2[tf2['Signal']==1]['Next Day Close'] - tf2[tf2['Signal']==1]['Next Day Open']).sum()
```

```python
# the statistics
get_stats(tf2['PnL'])
```

```python
# clearly, the strategy did not work well
# trying to improve the result by adding some restrictions on the signal
```

```python
def get_signal(r):
    if r['Predicted Next Close'] > r['Next Day Open'] + 1:
        return 1
    else:
        return 0
    
def get_ret(r):
    if r['Signal'] == 1:
        return ((r['Next Day Close'] - r['Next Day Open'])/r['Next Day Open']) * 100
    else:
        return 0
    
tf2 = tf2.assign(Signal = tf2.apply(get_signal, axis=1))
tf2 = tf2.assign(PnL = tf2.apply(get_ret, axis=1))
tf2
```

```python
# the return is improved
(tf2[tf2['Signal']==1]['Next Day Close'] - tf2[tf2['Signal']==1]['Next Day Open']).sum()
```

```python
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
```

```python
# splitting the close price into 5-elements series
sp['Close'].iloc[0:5]
```

```python
# computing the percentage changes for the first 4 elements of each time series and pairing them to
# the percentage change of the 5th element
```

```python
tseries = []
tlen = 5
for i in range(tlen, len(sp), tlen):
    pctc = sp['Close'].iloc[i-tlen:i].pct_change()[1:].values * 100
    res = sp['Close'].iloc[i-tlen:i+1].pct_change()[-1] * 100
    tseries.append((pctc, res))
```

```python
tseries[0]
```

```python
# computing the distance -  or similatities- between each two time series
def dtw_dist(x, y):
    distance, path = fastdtw(x, y, dist=euclidean)
    return distance
```

```python
dist_pairs = []
for i in range(len(tseries)):
    for j in range(len(tseries)):
        dist = dtw_dist(tseries[i][0], tseries[j][0])
        dist_pairs.append((i,j,dist,tseries[i][1], tseries[j][1]))
```

```python
dist_pairs[0:7]
```

```python
# creating a frame of the data
# deleting the zero-dist (the same series)
dist_frame = pd.DataFrame(dist_pairs, columns=['A','B','Dist', 'A Ret', 'B Ret'])
sf = dist_frame[dist_frame['Dist']>0].sort_values(['A','B']).reset_index(drop=1)
sfe = sf[sf['A']<sf['B']]
```

```python
# keeping the optimal dist -lest than one-
# keeping only the positive return of series A - profitable- 
winf = sfe[(sfe['Dist']<=1)&(sfe['A Ret']>0)]
winf
```
